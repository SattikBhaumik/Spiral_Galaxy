```python
import matplotlib.pyplot as plt
import numpy as np

# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Generate a list of prime numbers
prime_numbers = [p for p in range(2, 100) if is_prime(p)]

# Convert prime numbers to radians for theta
theta = [np.deg2rad(p) for p in prime_numbers]  # Though p is not in degrees, this line is for demonstration

# Using prime numbers for r
r = prime_numbers

# Create polar plot
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.scatter(theta, r)

# Set the display of the plot
ax.set_title("Prime Numbers in Polar Coordinates (r, theta) = (p, p)")

plt.show()
```

```python
# Offset each spiral by 60 degrees (pi/3 radians)
offsets = [i * (np.pi/3) for i in range(6)]

# Create polar plot with 6 spirals
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

for offset in offsets:
    # Adjust theta for each spiral
    theta_offset = [theta_i + offset for theta_i in theta]
    ax.scatter(theta_offset, r, c='red')

# Set the display of the plot
ax.set_title("Six Spirals with Prime Numbers in Polar Coordinates")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def spiral_galaxy(phi, A=1, B=0.5, N = 4):
    """
    Generate the coordinates for a spiral galaxy using the given formula.

    Parameters:
    - phi: Azimuthal angle in radians.
    - A, B, N: Parameters controlling the shape of the spiral.

    Returns:
    - r: Radius from the center of the galaxy to a point on the arm.
    """
    r = A / np.log(B * np.tan(phi / (2 * N)))
    return r

def is_prime(n):
    if n % 2 == 0 and n > 2:
        return False
    return all(n % i for i in range(3, int(math.sqrt(n)) + 1, 2))

# Generate a spiral galaxy
phi = np.linspace(0, 2 * np.pi, 1000)  # Generate values for phi
foo = np.vectorize(is_prime)
pbools = foo(phi)
primes = np.extract(pbools, phi)


r = spiral_galaxy(primes)  # Calculate radius for each phi

# Convert polar coordinates to Cartesian coordinates for plotting
x = r * np.cos(primes)
y = r * np.sin(primes)

# Plotting
plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.plot(-x, -y)  # Plotting the opposite arm
plt.title("Spiral Galaxy Simulation")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Reusing the spiral galaxy function from the previous plot
def spiral_galaxy(phi, A=1, B=0.5, N=4):
    r = A / np.log(B * np.tan(phi / (2 * N)))
    return r

# Function to check if a number is prime
def is_prime(n):
    if n % 2 == 0 and n > 2:
        return False
    return all(n % i for i in range(3, int(math.sqrt(n)) + 1, 2))

# Generate a spiral galaxy
phi = np.linspace(1, 100, 1000)  # Generate values for phi
foo = np.vectorize(is_prime)
pbools = foo(phi)
primes = np.extract(pbools, phi)

r = spiral_galaxy(primes)  # Calculate radius for each phi

# Create polar plot for the prime spirals
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.scatter(primes, r, c='red')  # Plotting primes on the spiral

# Set the display of the plot
ax.set_title("Spiral Galaxy with Prime Numbers in Polar Coordinates")
plt.show()
```
