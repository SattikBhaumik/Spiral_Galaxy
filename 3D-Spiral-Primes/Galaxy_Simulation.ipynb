```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Galaxy parameters
num_stars = 10000  # Number of stars in the galaxy
galaxy_radius = 50.0  # Radius of the galaxy
bulge_radius = 10.0  # Bulge radius
disk_thickness = 1.0  # Disk thickness
halo_radius = 100.0  # Halo radius

# Star distribution functions

def random_in_bulge(num_stars, bulge_radius):
    """
    Generate stars randomly within a spherical bulge.
    """
    r = bulge_radius * np.random.random(num_stars) ** (1/3)
    theta = np.random.uniform(0, np.pi, num_stars)
    phi = np.random.uniform(0, 2 * np.pi, num_stars)

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return np.array([x, y, z]).T

def random_in_disk(num_stars, galaxy_radius, disk_thickness):
    """
    Generate stars randomly within a rotating disk.
    """
    r = galaxy_radius * np.sqrt(np.random.random(num_stars))
    theta = np.random.uniform(0, 2 * np.pi, num_stars)
    z = np.random.normal(0, disk_thickness, num_stars)

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    return np.array([x, y, z]).T

def random_in_halo(num_stars, halo_radius):
    """
    Generate stars randomly within a spherical halo.
    """
    r = halo_radius * np.random.random(num_stars) ** (1/3)
    theta = np.random.uniform(0, np.pi, num_stars)
    phi = np.random.uniform(0, 2 * np.pi, num_stars)

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return np.array([x, y, z]).T

# Combine the star components
bulge_stars = random_in_bulge(int(num_stars * 0.1), bulge_radius)
disk_stars = random_in_disk(int(num_stars * 0.8), galaxy_radius, disk_thickness)
halo_stars = random_in_halo(int(num_stars * 0.1), halo_radius)

# Combine into one galaxy
stars = np.concatenate([bulge_stars, disk_stars, halo_stars])

# 3D Plotting with interactivity

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('black')
ax.grid(False)

# Color for the stars
star_colors = np.concatenate([['yellow']*len(bulge_stars),
                              ['white']*len(disk_stars),
                              ['blue']*len(halo_stars)])

# Scatter plot for the stars
ax.scatter(stars[:, 0], stars[:, 1], stars[:, 2], c=star_colors, s=0.1)

# Axis limits
ax.set_xlim([-halo_radius, halo_radius])
ax.set_ylim([-halo_radius, halo_radius])
ax.set_zlim([-halo_radius, halo_radius])

# Hide the axes
ax.set_axis_off()

# Interactive plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Galaxy parameters
num_stars = 100000
galaxy_radius = 50.0
bulge_radius = 10.0
disk_thickness = 1.0
halo_radius = 100.0
num_arms = 4
spiral_tightness = 0.3

# Star distribution functions

def random_in_bulge(num_stars, bulge_radius):
    """
    Generate stars randomly within a spherical bulge.
    """
    r = bulge_radius * np.random.random(num_stars) ** (1/3)
    theta = np.random.uniform(0, np.pi, num_stars)
    phi = np.random.uniform(0, 2 * np.pi, num_stars)

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return np.array([x, y, z]).T

def random_in_spiral_disk(num_stars, galaxy_radius, disk_thickness, num_arms, spiral_tightness):
    """
    Generate stars in a spiral disk pattern.
    """
    r = galaxy_radius * np.sqrt(np.random.random(num_stars))  # Radial distance
    arm_offset = np.random.choice(np.arange(num_arms), num_stars) * (2 * np.pi / num_arms)  # Spiral arm offset
    theta = spiral_tightness * np.log(r + 1) + arm_offset  # Logarithmic spiral equation

    # Add noise to the angular position
    theta += np.random.normal(0, 0.2, num_stars)

    z = np.random.normal(0, disk_thickness, num_stars)  # Disk thickness

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    return np.array([x, y, z]).T

def random_in_halo(num_stars, halo_radius):
    """
    Generate stars randomly within a spherical halo.
    """
    r = halo_radius * np.random.random(num_stars) ** (1/3)
    theta = np.random.uniform(0, np.pi, num_stars)
    phi = np.random.uniform(0, 2 * np.pi, num_stars)

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return np.array([x, y, z]).T

# Combine the star components
bulge_stars = random_in_bulge(int(num_stars * 0.1), bulge_radius)
disk_stars = random_in_spiral_disk(int(num_stars * 0.8), galaxy_radius, disk_thickness, num_arms, spiral_tightness)
halo_stars = random_in_halo(int(num_stars * 0.1), halo_radius)

# Combine into one galaxy
stars = np.concatenate([bulge_stars, disk_stars, halo_stars])

# 3D Plotting with visible axes, grid, and labels

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the stars
ax.scatter(stars[:, 0], stars[:, 1], stars[:, 2], c='white', s=0.1)

# Set axis labels
ax.set_xlabel('X Axis (kpc)')
ax.set_ylabel('Y Axis (kpc)')
ax.set_zlabel('Z Axis (kpc)')

# Set axis limits
ax.set_xlim([-halo_radius, halo_radius])
ax.set_ylim([-halo_radius, halo_radius])
ax.set_zlim([-halo_radius, halo_radius])

# Set axis grid to be visible and have ticks
ax.grid(True)

# Set background color
ax.set_facecolor('black')

# Set tick parameters (size of the ticks)
ax.tick_params(axis='x', colors='white', which='both')
ax.tick_params(axis='y', colors='white', which='both')
ax.tick_params(axis='z', colors='white', which='both')

# Set axis color
ax.w_xaxis.line.set_color('white')
ax.w_yaxis.line.set_color('white')
ax.w_zaxis.line.set_color('white')

# Set title of the plot
ax.set_title('3D Spiral Galaxy Simulation', color='white')

# Interactive plot with axes visible and grid
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Galaxy parameters
num_stars = 500000  # Total number of stars
galaxy_radius = 50.0  # Maximum radius of the galaxy
bulge_radius = 10.0  # Radius of the bulge
disk_thickness = 1.0  # Thickness of the disk
halo_radius = 100.0  # Halo radius
num_arms = 4  # Number of spiral arms

# Spiral formula parameters
A = 15  # Scale parameter
B = 1  # Spiral shape parameter
N = 4  # Determines spiral tightness

# Star distribution functions

def random_in_bulge(num_stars, bulge_radius):
    """
    Generate stars randomly within a spherical bulge.
    """
    r = bulge_radius * np.random.random(num_stars) ** (1/3)
    theta = np.random.uniform(0, np.pi, num_stars)
    phi = np.random.uniform(0, 2 * np.pi, num_stars)

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return np.array([x, y, z]).T

def random_in_spiral_disk(num_stars, galaxy_radius, disk_thickness, num_arms, A, B, N):
    """
    Generate stars in a spiral disk pattern using the given formula.
    """
    phi = np.random.uniform(0, 2 * np.pi, num_stars)  # Angular positions
    arm_offset = np.random.choice(np.arange(num_arms), num_stars) * (2 * np.pi / num_arms)  # Spiral arm offset
    phi += arm_offset  # Offset to create arms

    # Apply the spiral formula to determine r (radial distance)
    r = A / np.log(B * np.tan(phi / (2 * N)))

    # Apply some noise to simulate natural irregularities
    r += np.random.normal(0, 1, num_stars)

    # Convert to Cartesian coordinates
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    z = np.random.normal(0, disk_thickness, num_stars)  # Disk thickness (z-distribution)

    return np.array([x, y, z]).T

def random_in_halo(num_stars, halo_radius):
    """
    Generate stars randomly within a spherical halo.
    """
    r = halo_radius * np.random.random(num_stars) ** (1/3)
    theta = np.random.uniform(0, np.pi, num_stars)
    phi = np.random.uniform(0, 2 * np.pi, num_stars)

    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    return np.array([x, y, z]).T

# Combine the star components
bulge_stars = random_in_bulge(int(num_stars * 0.1), bulge_radius)
disk_stars = random_in_spiral_disk(int(num_stars * 0.8), galaxy_radius, disk_thickness, num_arms, A, B, N)
halo_stars = random_in_halo(int(num_stars * 0.1), halo_radius)

# Combine into one galaxy
stars = np.concatenate([bulge_stars, disk_stars, halo_stars])

# 3D Plotting with visible axes, grid, and labels

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the stars
ax.scatter(stars[:, 0], stars[:, 1], stars[:, 2], c='white', s=0.1)

# Set axis labels
ax.set_xlabel('X Axis (kpc)')
ax.set_ylabel('Y Axis (kpc)')
ax.set_zlabel('Z Axis (kpc)')

# Set axis limits
ax.set_xlim([-halo_radius, halo_radius])
ax.set_ylim([-halo_radius, halo_radius])
ax.set_zlim([-halo_radius, halo_radius])

# Set axis grid to be visible and have ticks
ax.grid(True)

# Set background color
ax.set_facecolor('black')

# Set tick parameters (size of the ticks)
ax.tick_params(axis='x', colors='white', which='both')
ax.tick_params(axis='y', colors='white', which='both')
ax.tick_params(axis='z', colors='white', which='both')

# Set axis color
ax.w_xaxis.line.set_color('white')
ax.w_yaxis.line.set_color('white')
ax.w_zaxis.line.set_color('white')

# Set title of the plot
ax.set_title('3D Spiral Galaxy Simulation (Custom Formula)', color='white')

# Interactive plot with axes visible and grid
plt.show()
```
